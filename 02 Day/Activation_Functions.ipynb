```python
# parse data
# skip header
lines = data_str.split('\n')[1:]

data = [lines.split(',') for lines in lines]
```

```python
X = np.array([[float(val) for val in row[:4]] for row in data])
y_str = [row[4].strip('"') for row in data] # find the cat...
```

```python
# find the unique class ['setosa',versicolor,'verginica']
unique_classes = list(set(y_str))
print(unique_classes)
```

```python
class_to_int = {c : i for i ,c in enumerate(unique_classes)}
y = np.array([class_to_int[c] for c in y_str])
```

```python
# shuffle and split ( set seed)
np.random.seed(42)
indices = np.arange(X.shape[0])
np.random.shuffle(indices)
X = X[indices]
y = y[indices]
```

```python
split = int(0.8 * len(X)) # 120 Training data
```

```python
X_train = torch.tensor(X[:split], dtype=torch.float32)
y_train = torch.tensor(y[:split], dtype=torch.long)

X_test = torch.tensor(X[split:], dtype=torch.float32)
y_test = torch.tensor(y[split:], dtype=torch.long)
```

```python
import torch.nn as nn

class SimpleNN(nn.Module):
  def __init__(self,actication=None):
    super(SimpleNN,self).__init__()
    # input to hidden
    self.fc1 = nn.Linear(4,10)
    self.activation = actication
    self.fc2 = nn.Linear(10,3) # hidden to output 3 class

  def forward(self,x):
    x = self.fc1(x)
    if self.activation:
        x = self.activation(x) # apply activation
    return x
```

```python
import torch.optim as optim

torch.manual_seed(42)

def train_and_evaluate(activation):
  model = SimpleNN(activation)
  criterion = nn.CrossEntropyLoss()
  optimizer = optim.SGD(model.parameters(), lr=0.01)
  for epoch in range(100):
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs,y_train)
    loss.backward()
    optimizer.step()

  # Evaluate
  with torch.no_grad():
    outputs = model(X_test)
    predicted = torch.argmax(outputs,dim=1)
    accuracy = (predicted == y_test).float().mean().item()
  return accuracy
```
