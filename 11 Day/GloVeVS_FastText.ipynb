```python
import gensim.downloader as api
```

```python
glove =api.load("glove-wiki-gigaword-300") # wikipedia + gigaword
```

```python
# wikipedia and english fast text
fast =api.load("fasttext-wiki-news-subwords-300")
```

```python
# interdface
print(glove.most_similar("king"))
print(fast.most_similar("king"))
```

```python
# analogy
print(glove.most_similar(positive=["king","woman"],negative=["man"]))
print(fast.most_similar(positive=["king","woman"],negative=["man"]))
```

```python
# unseen word
print(fast.most_similar(positive=("king")))
print(glove.most_similar(positive=("king")))
```
